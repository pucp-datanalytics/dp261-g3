# Ratio de utilizacion de credito

avg_bill: Promedio de deuda mensual del cliente

credit_utilization:
Mide qué porcentaje del crédito disponible está usando el cliente.

0.1 → usa poco crédito

0.5 → uso moderado

0.9+ → alto riesgo de sobreendeudamiento

```python
#Ratio de utilizacion de credito
df["avg_bill"] = df[[f"BILL_AMT{i}" for i in range(1,7)]].mean(axis=1)

df["credit_utilization"] = df["avg_bill"] / df["LIMIT_BAL"]
```

# Ratio de pago
avg_pay: Promedio de pagos mensuales del cliente.

payment_ratio: Cuanto paga vs cuanto debe.

```python
# Ratio de pago
df["avg_pay"] = df[[f"PAY_AMT{i}" for i in range(1,7)]].mean(axis=1)

df["payment_ratio"] = df["avg_pay"] / (df["avg_bill"] + 1)
```

# Riesgo de impago
late_payments: cuantos meses el cliente se retraso en pagos

```python
# Riesgo de impago
pay_cols = [col for col in df.columns if "PAY_" in col and col != "PAY_0"]
df["late_payments"] = df[pay_cols].apply(lambda x: (x > 0).sum(), axis=1)
```

bill_trend: Cambio de deuda entre el inicio y el final del periodo observado

Positivo deuda aumento y negativo la deuda disminuyo

```python
# Cambio de deuda
df["bill_trend"] = df["BILL_AMT6"] - df["BILL_AMT1"]
```

# Evolución de los pagos del cliente en el tiempo

```python
# Cambio de pagos 
df["pay_trend"] = df["PAY_AMT6"] - df["PAY_AMT1"]
```

```python
# Edad en grupos 
df["age_group"] = pd.cut(df["AGE"], bins=[0,25,35,50,100], labels=["young","adult","middle","senior"])
```

```python
# Credito por niveles
df["credit_level"] = pd.cut(df["LIMIT_BAL"], bins=5)
```

# Score de riesgo financiero simple

LIMIT_BAL: Cuanto credito tiene el cliente

late_payments: Cuantas veces se atraso


```python
# Riesgo financiero
df["risk_score"] = df["LIMIT_BAL"] * df["late_payments"]
```

# Capacidad de pago del cliente

```python
# Capacidad de pago
df["ability_to_pay"] = df["PAY_AMT1"] + df["PAY_AMT2"] + df["PAY_AMT3"]
```

Transformacion de linea de credito evita valores grandes 

```python
df["log_limit"] = np.log1p(df["LIMIT_BAL"])
df["log_avg_bill"] = np.log1p(df["avg_bill"].clip(lower=0))
```

# Eliminacion por correlacion 

```python
corr = df.corr(numeric_only=True)
```

```python
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
```

```python
to_drop = [col for col in upper.columns if any(upper[col].abs() > 0.95)]
to_drop
```

```python
#Se eliminaron variables con correlación mayor a 0.95 para reducir multicolinealidad y mejorar estabilidad del modelo.
```

```python
df_reduced = df.drop(columns=to_drop)
```
