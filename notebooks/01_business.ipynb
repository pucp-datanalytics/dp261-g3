```python
Gestión de Riesgo Crediticio
Predicción de clientes de tarjeta de crédito que caerán en mora el próximo mes, para apoyar decisiones de gestión de cartera en banca/fintech
```

```python
Contexto de la Industria

El sector bancario y fintech enfrenta un desafío constante: la morosidad en tarjetas de crédito. Los incumplimientos de pago generan pérdidas financieras, deterioran la cartera y afectan la salud del sistema.

El dataset utilizado proviene de 30,000 clientes de tarjeta de crédito en Taiwán (abril–septiembre 2005), publicado en el UCI Machine Learning Repository por Yeh & Lien (2009). Incluye variables demográficas, límite de crédito, historial de pagos de los últimos seis meses, montos facturados y montos pagados.

En la actualidad (2025–2026), los bancos están incorporando modelos de machine learning para la gestión de riesgo crediticio. Asimismo, se prioriza la interpretabilidad de los modelos debido a regulaciones de transparencia en fintech. Finalmente, la integración de técnicas de IA generativa y modelos explicables (Explainable AI, XAI) se utiliza para justificar decisiones ante clientes y reguladores, fortaleciendo la confianza en los sistemas de predicción.
```

```python
**Problema a resolver**

El banco necesita identificar con anticipación qué clientes tienen mayor probabilidad de incumplir su pago mensual, para tomar acciones preventivas: ajustar límites de crédito, ofrecer planes de pago o priorizar gestión de cobranza.
```

```python

**Variable objetivo (target)**

- default.payment.next.month: variable binaria (0 = no mora, 1 = mora).
```

```python
**KPIs de éxito del modelo**

- Métrica principal: AUC-ROC (capacidad de discriminación).
- Umbral mínimo aceptable: ≥ 0.70.
- Métricas complementarias: Recall (para minimizar falsos negativos), F1-score (balance entre precisión y recall).
```

```python
**Glosario de términos**

- MORA: incumplimiento de pago en la fecha acordada.
- DEFAULT PAYMENT: variable binaria que indica si el cliente cayó en mora el mes siguiente.
- AUC-ROC: métrica que mide capacidad de clasificación.
- PAY 0 a PAY 6 :  Estado de pago mensual (-1=puntual, 1=1 mes de atraso, ..)
- BILL AMT1--6 : Monto facturado cada mes (NT$)

- PAY AMT1--6: Monto pagado cada mes (NT$)

- LIMIT BAL : Límite de credito otorgado (NT$)
```
